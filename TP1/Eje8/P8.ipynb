```python
import re
import os
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
%matplotlib inline
import math
from matplotlib import pyplot as plt
tokens_line =[]
with open("terminos.txt") as archivo:
    for linea in archivo:
        tokens_line.append((re.sub(r'[^0-9a-z]','',linea)))


terminos =tokens_line.index('terminos')
cant_line= tokens_line[terminos +1]
tokens_line.remove("terminos")
tokens_line.remove(cant_line)

def heaps_law (n, k, beta):
    return k*(n**beta)
```

```python
x = np.array([a for a in range(1,int(cant_line)+1)])
y = [int(elemento) for elemento in tokens_line]
plt.plot(x, y, linestyle='solid')
plt.title('Ley de Heaps')
plt.xlabel('Lineas del archivo')  
plt.ylabel('Terminos')  
plt.show()
```

```python
popt, pcov = curve_fit(heaps_law,x,y)
```

```python
plt.plot(x, y, label='datos reales')
plt.plot(x, heaps_law(x,*popt), color='red', label = "ajuste")
plt.title('Ley de Heaps')
plt.xlabel('Lineas del archivo')  
plt.ylabel('Terminos')  
plt.show()
```

```python
valores_parametros = [(1, 0.5), (1, 0.8), (1, 1.0), (1, 1.2)]

for i, parametros in enumerate(valores_parametros, start=1):
    plt.subplot(2, 2, i)
    plt.plot(x, y, linestyle='solid', label='Datos reales')
    plt.plot(x, heaps_law(x, *parametros), color='red', label=f'Ajuste (k={parametros[0]}, beta={parametros[1]})')
    plt.title(f'Ajuste {i}')
    plt.xlabel('Lineas del archivo')
    plt.ylabel('Terminos')
    plt.legend()

plt.tight_layout()
plt.show()
```
