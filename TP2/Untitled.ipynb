```python
def get_arr(file):
    arr = []
    for line in file:
        line = line.lower()
        line = line.replace('id','')
        line = line.replace('keywords','')
        line = line.replace('\n','')
        line = re.sub(r'[^a-z0-9(),:]','', line)
        ar =  line.split(":")
        if len(ar) > 1:
            doc_array = ar[1].strip('()').split(',')
            arr.append(doc_array) 
    return arr

    
```

```python
def m_booleano(doc,queri):
    mb = []
    for q in queri:
        if q in doc:
            mb.extend('1')
        else:
            mb.extend('0')
    if np.all(mb == '1'):
        print(mb)
    return mb
```
